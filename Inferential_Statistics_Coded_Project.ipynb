**Problem 2**

The breaking strength of gunny bags used for packaging cement is normally distributed with a mean of 5 kg per sq. centimeter and a standard deviation of 1.5 kg per sq. centimeter. The quality team of the cement company wants to know the following about the packaging material to better understand wastage or pilferage within the supply chain; Answer the questions below based on the given information; (Provide an appropriate visual representation of your answers, without which marks will be deducted)

2.1 What proportion of the gunny bags have a breaking strength of less than 3.17 kg per sq cm?

2.2 What proportion of the gunny bags have a breaking strength of at least 3.6 kg per sq cm.?

2.3 What proportion of the gunny bags have a breaking strength between 5 and 5.5 kg per sq cm.?

2.4 What proportion of the gunny bags have a breaking strength NOT between 3 and 7.5 kg per sq cm.?

```python
# Given data
mean = 5  # kg per sq. cm
std_dev = 1.5  # kg per sq. cm

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define the range of values for the normal distribution curve
x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000)
pdf = norm.pdf(x, mean, std_dev)

# Create plots
fig, axs = plt.subplots(2, 2, figsize=(14, 12))
plt.subplots_adjust(wspace=0.3, hspace=0.3)

# 2.1 Proportion with breaking strength < 3.17 kg
x1 = np.linspace(mean - 4*std_dev, 3.17, 1000)
pdf1 = norm.pdf(x1, mean, std_dev)
proportion_2_1 = norm.cdf(3.17, mean, std_dev)
axs[0, 0].plot(x, pdf, label='Normal Distribution', color='blue')
axs[0, 0].fill_between(x1, 0, pdf1, color='skyblue', alpha=0.4)
axs[0, 0].axvline(3.17, color='red', linestyle='--', label='x = 3.17 kg')
axs[0, 0].set_title(f'Proportion < 3.17 kg: {proportion_2_1:.4f}')
axs[0, 0].legend()
axs[0, 0].set_xlabel('Breaking Strength (kg per sq cm)')
axs[0, 0].set_ylabel('Probability Density')

# 2.2 Proportion with breaking strength >= 3.6 kg
x2 = np.linspace(3.6, mean + 4*std_dev, 1000)
pdf2 = norm.pdf(x2, mean, std_dev)
proportion_2_2 = 1 - norm.cdf(3.6, mean, std_dev)
axs[0, 1].plot(x, pdf, label='Normal Distribution', color='blue')
axs[0, 1].fill_between(x2, 0, pdf2, color='lightgreen', alpha=0.4)
axs[0, 1].axvline(3.6, color='red', linestyle='--', label='x = 3.6 kg')
axs[0, 1].set_title(f'Proportion >= 3.6 kg: {proportion_2_2:.4f}')
axs[0, 1].legend()
axs[0, 1].set_xlabel('Breaking Strength (kg per sq cm)')
axs[0, 1].set_ylabel('Probability Density')

# 2.3 Proportion with breaking strength between 5 and 5.5 kg
x3 = np.linspace(5, 5.5, 1000)
pdf3 = norm.pdf(x3, mean, std_dev)
proportion_2_3 = norm.cdf(5.5, mean, std_dev) - norm.cdf(5, mean, std_dev)
axs[1, 0].plot(x, pdf, label='Normal Distribution', color='blue')
axs[1, 0].fill_between(x3, 0, pdf3, color='salmon', alpha=0.4)
axs[1, 0].axvline(5, color='red', linestyle='--', label='x = 5 kg')
axs[1, 0].axvline(5.5, color='orange', linestyle='--', label='x = 5.5 kg')
axs[1, 0].set_title(f'Proportion between 5 and 5.5 kg: {proportion_2_3:.4f}')
axs[1, 0].legend()
axs[1, 0].set_xlabel('Breaking Strength (kg per sq cm)')
axs[1, 0].set_ylabel('Probability Density')

# 2.4 Proportion with breaking strength NOT between 3 and 7.5 kg
proportion_2_4 = 1 - (norm.cdf(7.5, mean, std_dev) - norm.cdf(3, mean, std_dev))
x4_left = np.linspace(mean - 4*std_dev, 3, 1000)
x4_right = np.linspace(7.5, mean + 4*std_dev, 1000)
pdf4_left = norm.pdf(x4_left, mean, std_dev)
pdf4_right = norm.pdf(x4_right, mean, std_dev)
axs[1, 1].plot(x, pdf, label='Normal Distribution', color='blue')
axs[1, 1].fill_between(x4_left, 0, pdf4_left, color='lightcoral', alpha=0.4)
axs[1, 1].fill_between(x4_right, 0, pdf4_right, color='lightcoral', alpha=0.4)
axs[1, 1].axvline(3, color='red', linestyle='--', label='x = 3 kg')
axs[1, 1].axvline(7.5, color='orange', linestyle='--', label='x = 7.5 kg')
axs[1, 1].set_title(f'Proportion NOT between 3 and 7.5 kg: {proportion_2_4:.4f}')
axs[1, 1].legend()
axs[1, 1].set_xlabel('Breaking Strength (kg per sq cm)')
axs[1, 1].set_ylabel('Probability Density')

plt.show()
```

The plots show the normal distribution curve with shaded areas representing the proportions for each question.

2.1: The shaded area to the left of 3.17 kg represents the proportion of bags with a breaking strength less than 3.17 kg.

2.2: The shaded area to the right of 3.6 kg represents the proportion with a breaking strength of at least 3.6 kg.

2.3: The shaded area between 5 kg and 5.5 kg represents the proportion of bags in that range.

2.4: The shaded areas outside the range of 3 kg to 7.5 kg represent the proportion of bags not within that range.

Shapiro-Wilk’s test
For validating the normality assumption, Shapiro-Wilk’s test is applied to the response variable. The Shapiro-Wilk test examines if the population from which the sample is drawn follows a normal distribution or not.

The null and alternative hypotheses can be stated as follows:

Null Hypothesis: The data follow a normal distribution.

Alternate Hypothesis: The data do not follow a normal distribution.

#### **Levene's Test for Homogeneity**

Levene's Test for Homogeneity of Variances
Hypotheses for Levene's Test:

Null Hypothesis (H0): The variances are equal across the groups.

Alternate Hypothesis (H1): The variances are not equal across the groups

```python
def perform_levene_test(df):
    print("\nLevene's Test for Homogeneity of Variances")
    groups = [df[df['Dentist'] == dentist]['Response'] for dentist in df['Dentist'].unique()]
    stat, p_value = stats.levene(*groups)
    print(f"Levene's test p-value = {p_value:.4f}")
```

```python
perform_levene_test(df)
```

```python
# alpha = 0.05
if p_value < alpha:
    print("The variances are equal across the groups (fail to reject H0).")
else:
    print("The variances are not equal across the groups (reject H0).")
```

**There is statistically significant evidence to suggest that the variances of the response (hardness of implants) are different between at least two dentists and hence we reject the Null Hypothesis**

We will use ANOVA (Analysis of Variance) to test if there are significant differences between the means of different methods. The assumptions for ANOVA are:

*   Normality: The hardness of implants should be approximately normally distributed for each method.
*   Homogeneity of Variances: The variances of hardness should be approximately equal across methods.



Conduct the hypothesis test and compute the p-value

We will use the scipy.stats library for ANOVA and Levene’s test. For pairwise comparison, we'll use Tukey's Honestly Significant Difference (HSD) test.

```python
import statsmodels.stats.multicomp as mc
df = pd.DataFrame(my_data)

# Perform separate analysis for each alloy
for alloy in df['Alloy'].unique():
    subset = df[df['Alloy'] == alloy]

# Check assumptions
# 1. Homogeneity of variances
_, p_levene = stats.levene(subset[subset['Method'] == 1]['Response'],
                               subset[subset['Method'] == 2]['Response'],
                               subset[subset['Method'] == 3]['Response'])
print(f"Levene's test p-value for Alloy {alloy}: {p_levene}")
perform_levene_test(subset)

# 2. ANOVA
model = ols('Response ~ C(Method)', data=subset).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print(f"ANOVA table for Alloy {alloy}:\n{anova_table}\n")

p_value_anova = anova_table['PR(>F)'][0]
print(f"ANOVA p-value for Alloy {alloy}: {p_value_anova}")

# If ANOVA is significant, perform Tukey's HSD test
if p_value_anova < 0.05:
    tukey = mc.pairwise_tukeyhsd(subset['Response'], subset['Method'])
    print(f"Tukey HSD test results for Alloy {alloy}:\n{tukey}\n")

# Check for normality
plt.figure(figsize=(12, 6))
for method in subset['Method'].unique():
    sns.histplot(subset[subset['Method'] == method]['Response'], kde=True, label=f'Method {method}')
    plt.title(f'Distribution of Responses for Alloy {alloy}')
    plt.legend()
    plt.show()

# Q-Q plot for normality
plt.figure(figsize=(12, 6))
for method in subset['Method'].unique():
    stats.probplot(subset[subset['Method'] == method]['Response'], dist="norm", plot=plt)
    plt.title(f'Q-Q Plot for Method {method} (Alloy {alloy})')
    plt.show()

# Note: The above code assumes that the data is properly cleaned and there are sufficient data points for each method and alloy.
```

### **4.4 How does the hardness of implants vary depending on dentists and methods together?**



For each alloy, the hypotheses for the ANOVA (Analysis of Variance) test are:

Null Hypothesis (H₀): The mean hardness of implants is the same across all methods.

Alternate Hypothesis (H₁): At least one method has a different mean hardness compared to the others.

The assumptions for ANOVA include:

Independence: Samples are independent.
Normality: Data within each group should be approximately normally distributed.
Homogeneity of Variances: Variances among the groups should be approximately equal

We will use ANOVA to test the hypothesis and post hoc tests to identify which pairs of methods differ.

```python
df = pd.DataFrame(my_data)

# Function to perform ANOVA and Tukey HSD test
def perform_anova_tukey(df, alloy):
    df_alloy = df[df['Alloy'] == alloy]
    methods = df_alloy['Method'].unique()

    # Collect data for each method
    data_groups = [df_alloy[df_alloy['Method'] == method]['Response'] for method in methods]

    # Perform ANOVA
    f_stat, p_value = stats.f_oneway(*data_groups)
    print(f"ANOVA results for Alloy {alloy}: F-statistic = {f_stat:.4f}, p-value = {p_value:.4f}")

    # Check assumptions
    print("\nAssumptions check:")
    for method, data in zip(methods, data_groups):
        _, p_norm = stats.shapiro(data)
        print(f"Method {method} - Shapiro-Wilk test for normality p-value: {p_norm:.4f}")

    # Tukey HSD test for pairwise comparisons
    from statsmodels.stats.multicomp import pairwise_tukeyhsd
    tukey = pairwise_tukeyhsd(df_alloy['Response'], df_alloy['Method'])
    print("\nTukey HSD results:")
    print(tukey)

    # Plot
    plt.figure(figsize=(10, 6))
    sns.boxplot(x='Method', y='Response', data=df_alloy)
    plt.title(f'Boxplot of Hardness for Alloy {alloy}')
    plt.show()

# Perform analysis for each alloy
for alloy in df['Alloy'].unique():
    perform_anova_tukey(df, alloy)
```
