# **Importing Library**

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = "Market_Risk_Data_coded.csv"
data = pd.read_csv('/content/Market_Risk_Data_coded.csv')
```

```python
# Convert 'Date' column to datetime format

data['Date'] = pd.to_datetime(data['Date'], format='%d-%m-%Y')
```

```python
# Set 'Date' as the index for time series plotting

data.set_index('Date', inplace=True)
```

```python
plt.figure(figsize=(14, 7))
```

```python
for column in data.columns:
    if pd.api.types.is_numeric_dtype(data[column]):
        plt.plot(data.index, data[column], label=column)
plt.title('Stock Price vs Time (Weekly Data)')
plt.xlabel('Date')
plt.ylabel('Stock Price (INR)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Plot stock prices over time

plt.figure(figsize=(14, 7))
for column in data.columns:
    plt.plot(data.index, data[column], label=column)
```

# **Stock Returns Calculation and Analysis**

```python
# Calculate weekly returns

returns = data.pct_change().dropna()
```

```python
# Calculate mean and standard deviation for each stock

mean_returns = returns.mean()
std_returns = returns.std()
```

```python
# Plot Mean vs Standard Deviation

plt.figure(figsize=(10, 6))
plt.scatter(std_returns, mean_returns, color='blue')
```

```python
# Annotate stock names

for stock in returns.columns:
    plt.annotate(stock, (std_returns[stock], mean_returns[stock]), textcoords="offset points", xytext=(5,5), ha='left')

# Set better limits to zoom in
plt.xlim(0, std_returns.max() + 0.01)
plt.ylim(mean_returns.min() - 0.001, mean_returns.max() + 0.001)

plt.title('Mean Return vs Standard Deviation (Volatility)')
plt.xlabel('Standard Deviation (Risk)')
plt.ylabel('Mean Weekly Return')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Print results

print("Mean Weekly Returns:\n", mean_returns)
print("\nStandard Deviation of Weekly Returns:\n", std_returns)
```
